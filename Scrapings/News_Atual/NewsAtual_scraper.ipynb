```python
def returnBsObj(url):
    link = urlopen(url)
    bsObj = BeautifulSoup(link, 'html.parser')
    return bsObj
```

```python
lista_title = []
lista_url = []
lista_texto = []
#198,108
```

```python
i=1
while True:
    
    url = r"https://www.newsatual.com/page/" + str(i) + r"/"
    try:
        bsObj = returnBsObj(url)
    except:
        break
    
    for item in bsObj.findAll('h2', {"class":"entry-title"}):
        
        link = item.find_next()['href']
        lista_url.append(link)
        
        lista_title.append(item.getText())
        
        page = returnBsObj(link)
        div = page.find('div',{'class':'entry-content clearfix'})
    
        texto = str()
        for text in div.findAll('p'):
            texto = texto + text.getText()
        lista_texto.append(texto)
    
    print(i)

    i+=1
```

```python
df = pd.DataFrame()
df['Título da Notícia'] = lista_title
df['URL'] = lista_url
df['Texto'] = lista_texto
```
